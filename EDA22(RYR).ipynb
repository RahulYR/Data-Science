```python
#Using StandardScaler & MinMaxScaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler_standard=StandardScaler()
scaler_minmax=MinMaxScaler()

#Apply scaling to numerical features.
num_features = f.select_dtypes(include=['int64', 'float64']).columns
f[num_features] = scaler_standard.fit_transform(f[num_features])
f[num_features] = scaler_minmax.fit_transform(f[num_features])
f[num_features]
```

```python
categorical_features = f.select_dtypes(include=['object']).columns
for col in categorical_features:
    f[col].fillna(f[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# One-Hot Encoding for categorical variables with less than 5 categories
one_hot_columns = [col for col in f.select_dtypes(include=['object']).columns if f[col].nunique() < 5]
f_one_hot_encoded = pd.get_dummies(f, columns=one_hot_columns)

# Label Encoding for categorical variables with more than 5 categories
label_encoder = LabelEncoder()
label_encoded_columns = [col for col in f.select_dtypes(include=['object']).columns if f[col].nunique() > 5]
f_label_encoded = f.copy()
for col in label_encoded_columns:
    f_label_encoded[col] = label_encoder.fit_transform(f_label_encoded[col])

f_one_hot_encoded.head(), f_label_encoded.head()
```

```python
# Creating new features
f['family_size'] = f['fnlwgt'] + f['education_num']  # Example combining fnlwgt and education_num
f['is_alone'] = (f['relationship'] == 'Not-in-family').astype(int)

# Applying log transformation to a skewed numerical feature
f['capital_gain_log'] = f['capital_gain'].apply(lambda x: np.log(x + 1))  # Adding 1 to avoid log(0)

f[['family_size', 'is_alone', 'capital_gain_log']].head()
```

```python
from sklearn.ensemble import IsolationForest

# Using Isolation Forest to detect outliers
iso = IsolationForest(contamination=0.1)
f['outlier'] = iso.fit_predict(f[num_features])

# Removing outliers
f_no_outliers = f[f['outlier'] != -1].drop(columns=['outlier'])

f_no_outliers.head()
```
